```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

```

```python
#leggiamo l'immagine dell'occhio
eye_image = cv2.imread('testCC.JPG', cv2.IMREAD_UNCHANGED)
medianCC = cv2.medianBlur(eye_image, 5)


SX1 = cv2.imread('testSX1.JPG', cv2.IMREAD_UNCHANGED)
medianSX = cv2.medianBlur(SX1, 5)


DX = cv2.imread('testDX.JPG', cv2.IMREAD_UNCHANGED)
medianDX = cv2.medianBlur(DX, 5)



compare = np.concatenate((cv2.resize(DX, (600,450)), cv2.resize(eye_image, (600,450)), cv2.resize(SX1, (600,450))), axis=1) #side by side comparison
compare2 = np.concatenate((cv2.resize(medianDX, (600,450)), cv2.resize(medianCC, (600,450)), cv2.resize(medianSX, (600,450))), axis=1) #side by side comparison
cv2_imshow(compare)
cv2_imshow(compare2)


```

```python
eye_image = cv2.imread('testCC.JPG', cv2.IMREAD_GRAYSCALE)
medianCC = cv2.medianBlur(eye_image, 5)

SX1 = cv2.imread('testSX1.JPG', cv2.IMREAD_GRAYSCALE)
medianSX = cv2.medianBlur(SX1, 5)

DX = cv2.imread('testDX.JPG', cv2.IMREAD_GRAYSCALE)
medianDX = cv2.medianBlur(DX, 5)


compare = np.concatenate((cv2.resize(DX, (600,450)), cv2.resize(eye_image, (600,450)), cv2.resize(SX1, (600,450))), axis=1) #side by side comparison
compare2 = np.concatenate((cv2.resize(medianDX, (600,450)), cv2.resize(medianCC, (600,450)), cv2.resize(medianSX, (600,450))), axis=1)
cv2_imshow(compare)
cv2_imshow(compare2)
```

```python
# Applica una soglia per ottenere un'immagine binaria
_, thresholded_eye = cv2.threshold(eye_image, 50, 255, cv2.THRESH_BINARY)
_, thresholded_medianCC = cv2.threshold(medianCC, 50, 255, cv2.THRESH_BINARY)

_, thresholded_SX1 = cv2.threshold(SX1, 50, 255, cv2.THRESH_BINARY)
_, thresholded_medianSX = cv2.threshold(medianSX, 50, 255, cv2.THRESH_BINARY)

_, thresholded_DX = cv2.threshold(DX, 50, 255, cv2.THRESH_BINARY)
_, thresholded_medianDX = cv2.threshold(medianDX, 50, 255, cv2.THRESH_BINARY)


compare = np.concatenate((cv2.resize(thresholded_DX, (600,450)), cv2.resize(thresholded_eye, (600,450)), cv2.resize(thresholded_SX1, (600,450))), axis=1) #side by side comparison
compare2 = np.concatenate((cv2.resize(thresholded_medianDX, (600,450)), cv2.resize(thresholded_medianCC, (600,450)), cv2.resize(thresholded_medianSX, (600,450))), axis=1) #side by side comparison
cv2_imshow(compare)
cv2_imshow(compare2)
```

```python
# Definisci il kernel per la dilatazione ed erosione
kernel = np.ones((5, 5), np.uint8)
```

```python
# Dilatazione
dilated_eye = cv2.dilate(thresholded_eye, kernel, iterations=1)
dilated_medianCC = cv2.dilate(thresholded_medianCC, kernel, iterations=1)

dilated_SX1 = cv2.dilate(thresholded_SX1, kernel, iterations=1)
dilated_medianSX = cv2.dilate(thresholded_medianSX, kernel, iterations=1)


dilated_DX = cv2.dilate(thresholded_DX, kernel, iterations=1)
dilated_medianDX = cv2.dilate(thresholded_medianDX, kernel, iterations=1)



compare = np.concatenate((cv2.resize(dilated_DX, (600,450)), cv2.resize(dilated_eye, (600,450)), cv2.resize(dilated_SX1, (600,450))), axis=1) #side by side comparison
compare2 = np.concatenate((cv2.resize(dilated_medianDX, (600,450)), cv2.resize(dilated_medianCC, (600,450)), cv2.resize(dilated_medianSX, (600,450))), axis=1) #side by side comparison
cv2_imshow(compare)
cv2_imshow(compare2)
```

```python
# Erosione
eroded_eye = cv2.erode(dilated_eye, kernel, iterations=1)
eroded_medianCC = cv2.erode(dilated_medianCC, kernel, iterations=1)

eroded_SX1 = cv2.erode(dilated_SX1, kernel, iterations=1)
eroded_medianSX = cv2.erode(dilated_medianSX, kernel, iterations=1)


eroded_DX = cv2.erode(dilated_DX, kernel, iterations=1)
eroded_medianDX = cv2.erode(dilated_medianDX, kernel, iterations=1)



compare = np.concatenate((cv2.resize(eroded_DX, (600,450)), cv2.resize(eroded_eye, (600,450)), cv2.resize(eroded_SX1, (600,450))), axis=1) #side by side comparison
compare2 = np.concatenate((cv2.resize(eroded_medianDX, (600,450)), cv2.resize(eroded_medianCC, (600,450)), cv2.resize(eroded_medianSX, (600,450))), axis=1) #side by side comparison
cv2_imshow(compare)
cv2_imshow(compare2)
```
