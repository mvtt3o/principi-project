```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

```

```python
#leggiamo l'immagine dell'occhio

CC = cv2.medianBlur(cv2.imread('NewCC.JPG', cv2.IMREAD_UNCHANGED), 5)

SX = cv2.medianBlur(cv2.imread('NewSX.JPG', cv2.IMREAD_UNCHANGED), 5)

DX = cv2.medianBlur(cv2.imread('NewDX.JPG', cv2.IMREAD_UNCHANGED), 5)

compare = np.concatenate((cv2.resize(DX, (600,450)), cv2.resize(CC, (600,450)), cv2.resize(SX, (600,450))), axis=1) #side by side comparison

cv2_imshow(compare)



```

```python
CC = cv2.medianBlur(cv2.imread('NewCC.JPG', cv2.IMREAD_GRAYSCALE), 5)

SX = cv2.medianBlur(cv2.imread('NewSX.JPG', cv2.IMREAD_GRAYSCALE), 5)

DX = cv2.medianBlur( cv2.imread('NewDX.JPG', cv2.IMREAD_GRAYSCALE), 5)

compare = np.concatenate((cv2.resize(DX, (600,450)), cv2.resize(CC, (600,450)), cv2.resize(SX, (600,450))), axis=1) #side by side comparison

cv2_imshow(compare)
```

```python
# Applica una soglia per ottenere un'immagine binaria
_, thresholded_CC = cv2.threshold(CC, 65, 255, cv2.THRESH_BINARY)

_, thresholded_SX = cv2.threshold(SX, 65, 255, cv2.THRESH_BINARY)

_, thresholded_DX = cv2.threshold(DX, 65, 255, cv2.THRESH_BINARY)

compare = np.concatenate((cv2.resize(thresholded_DX, (600,450)), cv2.resize(thresholded_CC, (600,450)), cv2.resize(thresholded_SX, (600,450))), axis=1) #side by side comparison

cv2_imshow(compare)
```

```python
# Definisci il kernel per la dilatazione ed erosione
kernel = np.ones((15, 15), np.uint8)
```

```python
# Dilatazione
dilated_CC = cv2.dilate(thresholded_CC, kernel, iterations=1)

dilated_SX = cv2.dilate(thresholded_SX, kernel, iterations=1)

dilated_DX = cv2.dilate(thresholded_DX, kernel, iterations=1)

compare = np.concatenate((cv2.resize(dilated_DX, (600,450)), cv2.resize(dilated_CC, (600,450)), cv2.resize(dilated_SX, (600,450))), axis=1) #side by side comparison

cv2_imshow(compare)
```

```python
# Erosione
eroded_CC = cv2.erode(dilated_CC, kernel, iterations=1)

eroded_SX = cv2.erode(dilated_SX, kernel, iterations=1)

eroded_DX = cv2.erode(dilated_DX, kernel, iterations=1)

compare = np.concatenate((cv2.resize(eroded_DX, (600,450)), cv2.resize(eroded_CC, (600,450)), cv2.resize(eroded_SX, (600,450))), axis=1) #side by side comparison

cv2_imshow(compare)
```
